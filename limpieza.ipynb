```python
import pandas as pd
from sklearn.impute import KNNImputer

# Cargar el archivo CSV
df = pd.read_csv(r'C:\Users\Danny\Desktop\NBA_database\csv\game_info_optimizada.csv')

# Definir las columnas que se van a imputar
columns_to_impute = ['attendance']

# Crear un DataFrame solo con las columnas a imputar
df_impute = df[columns_to_impute]

# Imputar valores faltantes utilizando KNNImputer
imputer = KNNImputer(n_neighbors=5)
df_imputed = imputer.fit_transform(df_impute)

# Convertir el resultado a un DataFrame
df_imputed = pd.DataFrame(df_imputed, columns=columns_to_impute)

# Reemplazar la columna 'attendance' en el DataFrame original
df['attendance'] = df_imputed['attendance']

# Redondear los valores de la columna 'attendance' a un decimal
df['attendance'] = df['attendance'].round(1)

# Mostrar el DataFrame después de la imputación
print("\nDataFrame después de la imputación con KNNImputer:")
print(df.head())
```

```python
import pyodbc

# Conectar a SQL Server
conn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER=100.115.35.29,1433\SQLEXPRESS;DATABASE=NBA_database;UID=danny1;PWD=1234')
cursor = conn.cursor()

# Crear las tablas
cursor.execute('''
CREATE TABLE games (
    game_id VARCHAR(20) PRIMARY KEY,
    game_date_est DATETIME,
    game_sequence INT
)
''')

cursor.execute('''
CREATE TABLE teams (
    team_id VARCHAR(20) PRIMARY KEY,
    team_abbreviation VARCHAR(10),
    team_city_name VARCHAR(50),
    team_nickname VARCHAR(50)
)
''')

cursor.execute('''
CREATE TABLE results (
    result_id INT PRIMARY KEY IDENTITY(1,1),
    game_id VARCHAR(20),
    team_id VARCHAR(20),
    is_home_team BIT,
    pts_qtr1 FLOAT,
    pts_qtr2 FLOAT,
    pts_qtr3 FLOAT,
    pts_qtr4 FLOAT,
    pts_ot1 FLOAT,
    pts_ot2 FLOAT,
    pts_ot3 FLOAT,
    pts_ot4 FLOAT,
    pts_ot5 FLOAT,
    pts_ot6 FLOAT,
    pts_ot7 FLOAT,
    pts_ot8 FLOAT,
    pts_ot9 FLOAT,
    pts_ot10 FLOAT,
    pts_total FLOAT,
    FOREIGN KEY (game_id) REFERENCES games(game_id),
    FOREIGN KEY (team_id) REFERENCES teams(team_id)
)
''')

cursor.execute('''
CREATE TABLE team_records (
    record_id INT PRIMARY KEY IDENTITY(1,1),
    team_id VARCHAR(20),
    game_id VARCHAR(20),
    wins_losses VARCHAR(10),
    FOREIGN KEY (team_id) REFERENCES teams(team_id),
    FOREIGN KEY (game_id) REFERENCES games(game_id)
)
''')

conn.commit()
```
