```python
import pandas as pd
from sklearn.impute import KNNImputer

# Cargar el archivo CSV
df = pd.read_csv(r'C:\Users\Danny\Desktop\NBA_database\csv\game_info_optimizada.csv')

# Definir las columnas que se van a imputar
columns_to_impute = ['attendance']

# Crear un DataFrame solo con las columnas a imputar
df_impute = df[columns_to_impute]

# Imputar valores faltantes utilizando KNNImputer
imputer = KNNImputer(n_neighbors=5)
df_imputed = imputer.fit_transform(df_impute)

# Convertir el resultado a un DataFrame
df_imputed = pd.DataFrame(df_imputed, columns=columns_to_impute)

# Reemplazar la columna 'attendance' en el DataFrame original
df['attendance'] = df_imputed['attendance']

# Redondear los valores de la columna 'attendance' a un decimal
df['attendance'] = df['attendance'].round(1)

# Mostrar el DataFrame después de la imputación
print("\nDataFrame después de la imputación con KNNImputer:")
print(df.head())
```

```python

data = pd.read_csv(r'C:\Users\Danny\Desktop\NBA_database\csv\game_summary.csv')
data.head()


# Eliminar columnas innecesarias
data = data.drop(columns=['game_sequence', 'live_pc_time', 'natl_tv_broadcaster_abbreviation', 'live_period_time_bcast'])

# Rellenar valores faltantes (en este caso, con un valor por defecto o eliminando filas)
data = data.fillna({'game_status_text': 'Unknown'})

# Convertir tipos de datos
data['game_date_est'] = pd.to_datetime(data['game_date_est'])
data['game_id'] = data['game_id'].astype(int)
data['home_team_id'] = data['home_team_id'].astype(int)
data['visitor_team_id'] = data['visitor_team_id'].astype(int)
data['season'] = data['season'].astype(int)
data['live_period'] = data['live_period'].astype(int)
data['wh_status'] = data['wh_status'].astype(int)

# Renombrar columnas
data = data.rename(columns={
    'game_date_est': 'game_date',
    'game_status_id': 'status_id',
    'game_status_text': 'status_text',
    'gamecode': 'game_code',
    'home_team_id': 'home_team',
    'visitor_team_id': 'visitor_team',
    'live_period': 'period',
    'wh_status': 'whistle_status'
})

# Eliminar duplicados
data = data.drop_duplicates()

```
