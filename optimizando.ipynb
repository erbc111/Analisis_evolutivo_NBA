```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('team_details.csv')

# Imputar valores faltantes en la columna 'arenacapacity' con la media de las capacidades existentes
mean_capacity = df['arenacapacity'].mean()
df['arenacapacity'].fillna(round(mean_capacity, 1), inplace=True)

# Rellenar manualmente el valor faltante en la columna 'headcoach' para el equipo Toronto Raptors
df.loc[df['abbreviation'] == 'TOR', 'headcoach'] = 'Darko Rajaković'  # Ejemplo de nombre de entrenador

# Guardar el DataFrame modificado en un nuevo archivo CSV
df.to_csv('team_details_cleaned.csv', index=False)

# Cargar y mostrar el nuevo archivo CSV
df_cleaned = pd.read_csv('team_details_cleaned.csv')
print(df_cleaned)
```

```python
# Cargar el archivo CSV
data = pd.read_csv('team_history.csv')

# Verificar los tipos de datos
print("Tipos de datos de cada columna:")
print(data.dtypes)
print("\n")

# Verificar valores únicos en cada columna
for column in data.columns:
    print(f"Valores únicos en la columna {column}:")
    print(data[column].unique())
    print("\n")

# Verificar si hay valores nulos
null_values = data.isnull().sum()
print("Valores nulos en cada columna:")
print(null_values)
print("\n")

# Verificar si hay duplicados
duplicates = data.duplicated().sum()
print(f"Número de filas duplicadas: {duplicates}")
```
